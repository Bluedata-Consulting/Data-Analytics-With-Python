```python
mobile_data=pd.read_csv("mobile_price-train_set.csv")
```

```python
mobile_data.head()
```

```python
mobile_data.isna().sum()
```

```python
#battery power across price range
plt.figure(figsize=(10,5)) #10 indicates the length and 5 indicates the width 
sns.barplot(x='price_range',y='battery_power',data=mobile_data)
plt.title("Battery power distribution by price range") #the heading of the graph
plt.xlabel("price range") #text on the x axis
plt.ylabel("battery power in mAh") #text on the y axis
plt.show() #will show the visualization
```

```python
#RAM across price range 
plt.figure(figsize=(10,5)) #10 indicates the length and 5 indicates the width 
sns.barplot(x='price_range',y='ram',data=mobile_data)
plt.title("RAM distribution by price range") #the heading of the graph
plt.xlabel("price range") #text on the x axis
plt.ylabel("RAM") #text on the y axis
plt.show() #will show the visualization
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='price_range',data=mobile_data)
plt.title("distribution of mobile price range")
plt.xlabel("price range")
plt.ylabel("count")
plt.show()
```

```python
#relation between RAM and price range
plt.figure(figsize=(10,5))
sns.boxplot(x='price_range',y='battery_power',data=mobile_data)
plt.title("Battery vs Price range")
plt.xlabel("Price range")
plt.ylabel("Battery power in mAH")
plt.show()
```

```python
spam_data=pd.read_csv("spam.csv",encoding='ISO-8859-1')
```

```python
spam_data.head()
```

```python
#remove the unwanted columns and rename the column names e.g. v1-label and v2-message
spam_cleaned=spam_data[['v1','v2']].rename(columns={'v1':'label','v2':'message'})
```

```python
spam_cleaned.head()
```

```python
spam_cleaned.info()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='label',data=spam_cleaned)
plt.title("Distribution of spam and ham messages")
plt.xlabel("message type")
plt.ylabel("count")
plt.show()
```

```python
#message length distribution
spam_cleaned['message_length']=spam_cleaned['message'].apply(len) #len() applied on message column
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=spam_cleaned,x='message_length',hue='label')
plt.title("Message length distribution")
plt.xlabel("message length")
plt.ylabel("frequency")
plt.show()
```

```python
from wordcloud import WordCloud
from collections import Counter
```

```python
def generate_wordcloud(text,title):
    wordcloud=WordCloud(width=800,height=400)
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud,interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

spam_messages=spam_cleaned[spam_cleaned['label']=='spam']['message']
ham_messages=spam_cleaned[spam_cleaned['label']=='ham']['message']

spam_text=" ".join(spam_messages)
ham_text=" ".join(ham_messages) 
```

```python
generate_wordcloud(spam_text,"Most common words in spam message")
```

```python
generate_wordcloud(ham_text,"most common words in ham messages")
```
