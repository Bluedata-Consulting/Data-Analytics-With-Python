```python
#dealing with JSON files

import json #json is a built in package to support JSON files in Python

#data to be written
dictionary={
    "name":"Meera",
    "roll_no":56,
    "cgpa":8.6,
    "age":13
}

#serializing json
json_object=json.dumps(dictionary,indent=4)

#writing to sample.json
with open("sample.json","w") as outfile:
    outfile.write("[")
    outfile.write(json_object)
    outfile.write("]")
```

```python
#reading the json file created 
json_file=pd.read_json("sample.json") 
print(json_file)
```

```python
#exporting the json file created 
json_file.to_json("DAWP_sample.json") #.to_json helps in exporting the json data created to our local systems
```

```python
#data manipulation - joining the datasets
raw_data={
    'subject_id':[1,2,3,4],
    'first_name':['Alex','Peter','Tom','Jerry'],
    'last_name':['Anderson','Aoni','Parker','Jackman'],
}

df_a=pd.DataFrame(raw_data)
df_a.set_index("subject_id",inplace=True)

raw_data={
    'subject_id':[3,4,5,6],
    'first_name':['Tom','Jerry','Allan','Catherine'],
    'last_name':['Parker','Jackman','John','Ackerman'],
}

df_b=pd.DataFrame(raw_data)
df_b.set_index("subject_id",inplace=True)
```

```python
df_a
```

```python
df_b
```

```python
#performing inner join on top of common subject_id column(key)
df_a.join(df_b,how='inner',lsuffix='_caller',rsuffix='_other')
```

```python
#data manipulation - merging the datasets
df_new=df_a.merge(df_b) #merging will merge two dataframes having unique columns 
```
